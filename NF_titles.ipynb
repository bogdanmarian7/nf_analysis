```python
# Create a separate dataframe for Movies in the dataset
movies_df = netflix_df[netflix_df['type']=='Movie']
movies_df.name = "Movies Dataframe"
# Create a separate dataframe for TV Shows in the dataset
tvshows_df = netflix_df[netflix_df['type']=="TV Show"]
tvshows_df.name = "TV Shows Dataframe"
print("""The entire Dataset has been split to two separate dataframes, one containing the movies ({}) and one containing the TV Shows ({})""".format(movies_df.name, tvshows_df.name))
```

```python
# Pie plot
fig = plt.figure(figsize = (14, 5))

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} titles)".format(pct, absolute)
data = [movies_df.title.count(), tvshows_df.title.count()]


plt.pie(data, labels= ['Movies', 'TV Shows'], autopct = lambda pct: func(pct, data))

plt.title("Distribution of Movies and TV shows on Netflix")
plt.show()
```

```python
def get_countries():
    countries_available_movies = movies_df.country.squeeze()
    countries = []
    for countries_available in countries_available_movies.values:
        countries_list = str(countries_available).split(",")
        for country in countries_list:
            if country not in countries:
                countries.append(country)
    countries = [country.strip() for country in countries]
    countries = [country for country in countries if len(country) > 0]
    countries.remove("nan")
    countries = list(set(countries))
    return countries
    
```

```python
countries = get_countries()
```

```python
def get_cnt_bycountry(country, df):
    titles_cnt = 0
    for country_clm in df.country:
        if (country in str(country_clm)) or ("nan" in str(country_clm)):
            titles_cnt+=1
    return titles_cnt
```

```python
get_cnt_bycountry('Romania', netflix_df)
```

```python
for country in countries:
    cnt = get_cnt_bycountry(country, netflix_df)
    print(f"Country {country} has {cnt} titles.")
    
```

```python
def count_titles(df, countries_list):
    titles_df = pd.DataFrame(columns=['Country', 'Titles_Count'])
    for country in countries_list:
        titles_count = 0
        for countries_item in df.country:
            if (country in str(countries_item)) or ("nan" in str(countries_item)):
                titles_count+=1
        titles_df.loc[len(titles_df)] = [country, titles_count]
        #titles_df = titles_df.append([country, titles_count], ignore_index = True)
        #print(f"Country {country} has {titles_count} titles.")
    return titles_df
        
result = count_titles(netflix_df, countries)
    
```

```python
max_titles_count_df = result.sort_values('Titles_Count', ignore_index=True, ascending=False).head(10)
```

```python
# Countries with the highest number of titles available on Netflix platform
# Bar plot

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i], ha = 'center')

countries= max_titles_count_df.Country
titles_count = max_titles_count_df.Titles_Count
fig = plt.figure(figsize = (14, 5))

plt.bar(countries, titles_count, color ='green',
        width = 0.5)

addlabels(countries, titles_count)

plt.xlabel("Country")
plt.ylabel("Titles available per country")
plt.title("Countries with the highest number of titles available on Netflix platform")
plt.show()
```

```python
# Countries with the lowest number of titles available on Netflix platform
# Bar plot

min_titles_count_df = result.sort_values('Titles_Count', ignore_index=True).head(10)

countries= min_titles_count_df.Country
titles_count = min_titles_count_df.Titles_Count
fig = plt.figure(figsize = (14, 5))
#  Bar plot
plt.bar(countries, titles_count, color ='orange',
        width = 0.5)

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i], ha='center')
addlabels(countries, titles_count)

plt.xlabel("Country")
plt.ylabel("Titles available per country")
plt.title("Countries with the lowest number of titles available on Netflix platform")
plt.show()
```

```python
def count_titles_added_per_year(df, years):
    titles_added_df = pd.DataFrame(columns=['Year', 'Titles_Added'])
    for year in years:
        titles_added= df[df.date_added_year == year].shape[0]
        titles_added_df.loc[len(titles_added_df)] = [year, titles_added]
        
    return titles_added_df
```

```python
titles_added_df = count_titles_added_per_year(netflix_df, years)
```

```python
titles_added_df = titles_added_df.iloc[:5]
```

```python
titles_added_df = titles_added_df.sort_values('Year', ignore_index=True, ascending=False)
titles_added_df
```

```python
# Number of titles added on Netflix per year (On the last 5 years)
# Bar plot


years= titles_added_df.Year
titles_added = titles_added_df.Titles_Added
fig = plt.figure(figsize = (14, 5))
#  Bar plot
plt.bar(years, titles_added, color ='blue',
        width = 0.5)

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i], ha='center')
addlabels(years, titles_added)

plt.xlabel("Year")
plt.ylabel("Titles added")
plt.title("Number of titles added on Netflix per year")
plt.show()
```
